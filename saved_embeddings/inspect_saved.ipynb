```python
import torch
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics import silhouette_score

mpl.rcParams['figure.dpi']= 500
sns.set_context("notebook")
sns.set(style="white")
```

```python
def plot_with_clusters(data, clusters_labels, pathname=None, savename=None):
    plt.tight_layout(w_pad=1.5, h_pad=3.5)
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=clusters_labels, palette=sns.color_palette("hls", len(set(clusters_labels))), s=5, edgecolor='black', linewidth=0.5)
    plt.legend(prop={'size': 4})
    if pathname and savename:
        plt.savefig('%s/%s.pdf' % (pathname, savename))
```

# Baron1

```python
baron1_node_embs = np.load('baron1/node_embs.npy')
baron1_umap = np.load('baron1/umap.npy')
baron1_clusters = np.load('baron1/clusters.npy')
```

```python
baron1_gt_df = pd.read_csv('baron1/Baron1_ground_truth.csv')
```

```python
baron1_gt_clusters = baron1_gt_df['assigned_cluster'].values
```

```python
adjusted_rand_score(baron1_gt_clusters, baron1_clusters)
```

```python
silhouette_score(baron1_umap, baron1_clusters)
```

```python
plot_with_clusters(baron1_umap, baron1_clusters)
```

```python
baron1_metadata = pd.read_csv('baron1/metadata.csv', header=None, names=['param', 'value'])
```

```python
baron1_metadata
```
