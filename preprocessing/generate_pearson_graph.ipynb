```python
import numpy as np
import pandas as pd
```

```python
hvg_df = pd.read_csv('HVG/ProjectName_HVG_500.csv')
```

```python
hvg_df
```

```python
num_hvg = hvg_df.shape[0]
```

```python
def save_pearson_neighs(rna, k, name, weights=False):
    pearson_corr = rna.drop(columns=['Unnamed: 0']).corr()
    nlargest = k + 1
    order = np.argsort(-pearson_corr.values, axis=1)[:, :nlargest]
    to_save = np.delete(order, 0, 1)
    
    if weights:
        with open('Pearson/%s_pearson_graph_hvg_%d_k%d_w.txt' % (name, num_hvg, k), 'w') as out:
            for i in range(to_save.shape[0]):
                for j in range(to_save.shape[1]):
                    out.write('%d %d %.5f\n' % (i, to_save[i][j], pearson_corr.values[i][to_save[i][j]]))
    else:
        with open('Pearson/%s_pearson_graph_hvg_%d_k%d.txt' % (name, num_hvg, k), 'w') as out:
            for i in range(to_save.shape[0]):
                for j in range(to_save.shape[1]):
                    out.write('%d %d\n' % (i, to_save[i][j])) 
```

```python
project_name = 'ProjectName'
```

```python
save_pearson_neighs(hvg_df, 10, project_name)
```
